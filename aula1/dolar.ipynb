```python
dados.columns = ['datas', 'codigo', 'tipo', 'simbolo', 'compra', 'venda', 'ns1', 'ns2']
```

```python
dados.head()
```

```python
dados = dados.rename(columns={'ns2': 'ns10'})
dados.head()
```

```python
dados1 = dados.drop(columns=['ns1','ns10'])
dados1.head(2)
```

Análise dos tipos de atributos:

object: strings

int64: inteiros

float64: reais

complex: complexos

```python
dados1.dtypes
```

```python
dados1['compra'] = dados1['compra'].apply(lambda x: x.replace(',','.'))
dados1.head()
```

```python
dados1['venda'] = dados1['venda'].apply(lambda x: x.replace(',','.'))
dados1.head()
```

```python
dados1.dtypes
```

```python
dados1['compra'] = dados1['compra'].astype(float)
dados1['venda'] = dados1['venda'].astype(float)
```

```python
dados1.dtypes
```

```python
# Alterar todos de uma vez
dados1[['compra', 'venda']] = dados1[['compra', 'venda']].apply(pd.to_numeric)
```

```python
# Salvando arquivo
dados1.to_csv('dados_tratados.csv', encoding = 'iso-8859-1', index = False)
```
