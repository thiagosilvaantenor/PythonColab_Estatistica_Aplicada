```python
# CRIANDO FUNÃÃO PROBABILIDADE
def probab (A, E):
  # A = obj de estudo, E = espaÃ§o amostral
  resultado = (A / E)*100
  print('{:.2f}'.format(resultado))
```

```python
# PROBABILIDADE DE RETIRAR PREFÃRENCIA COR PRETA
probab (len(cor_preta), len(dados))
```

```python
# PROBABILIDADE DE SER MENOR QUE 1.70m
probab (len(altura1), len(dados))
```

## **PROBABILIDADE** **DE NÃO OCORRER**

```python
# CRIANDO FUNÃÃO PROBABILIDADE DE NÃO OCORRER UM EVENTO
def probab_nao (A, E):
  resultado = (1- (A/E))*100
  print('{:.2f}'.format(resultado))
```

```python
cor_vermelho = dados.loc[dados.cor_favorita.isin(['vermelha', 'vermelho'])]
cor_vermelho.head()
```

```python
# PROBABILIDADE DE NÃO RETIRAR COR FAVORITA VERMELHO
probab_nao(len(cor_vermelho), len(dados))
```

## **Probabilidade da IntersecÃ§Ã£o de dois eventos independentes**

```python
# CRIANDO FUNÃÃO PROBABILIDADE DA INTERSECÃÃO DE DOIS EVENTOS INDEPENDENTES
def probab_inter(A,B, E):
  resultado = (A/E * B/E)*100
  print('{:.1f}'.format(resultado))
```

```python
# PROBABILIDADE DE SORTEAR DUAS PESSOAS SENDO UMA DELAS COM A ALTURA MENOR QUE 1,70
# E A OUTRA COM COR FAVORITA PRETA COM REPOSIÃÃO.
probab_inter(len(cor_preta),len(altura1),len(dados))
```

## Probabilidade da uniÃ£o (AUB)

```python
# FUNÃÃO
def probab_uniao(A,B,AB, E):
  resultado = (A/E + B/E - AB/E) * 100
  print('{:.2f}'.format(resultado))
```

```python
#PROBABILIDADE DE SELECIONAR
probab_uniao(len(cor_preta),len(altura1),(len(cor_preta)/len(dados)*len(altura1)/len(dados)), len(dados))
```
