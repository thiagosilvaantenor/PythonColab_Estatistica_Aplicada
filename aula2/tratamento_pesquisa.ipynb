## **TRATAMENTO DOS DADOS**

```python
# Excluir coluna
dados1 = dados.drop(columns=['Carimbo de data/hora'])
dados1.head(3)
```

```python
# Renomear colunas
dados1 = dados1.rename(columns={'Qual a sua altura?': 'altura'})
dados1.head(2)
```

```python
dados1.rename(columns={'Qual o seu RA?': 'ra','Qual a sua cor favorita?':'cor_favorita',
                      'Qual o número de seu calçado?':'num_calcado'},inplace=True)
dados1.head(2)
```

```python
# Valores faltantes (missing)
dados1.isnull().sum()
```

```python
# Analisando as variáveis
dados1.dtypes
```

```python
dados1.head(23)
```

```python
dados1['altura'] = dados1['altura'].apply(lambda x: x.replace(',','.'))
dados1
```

```python
dados1.dtypes
```

```python
dados1['altura'] = dados1['altura'].astype(float)
dados1.dtypes
```

```python
dados1['cor_favorita'] = dados1.cor_favorita.str.lower()
dados1.head(23)
```

```python
import plotly.express as px
bolhas = px.scatter(dados1, x="num_calcado", y="altura")
bolhas.show()
```

```python
dados1['num_calcado'] = dados1['num_calcado'].apply(lambda x: x.replace('40/41', '41'))
dados1['num_calcado'] = dados1['num_calcado'].apply(lambda x: x.replace('43-44', '44'))
```

```python
dados1['num_calcado'] = dados1['num_calcado'].astype(int)
dados1.dtypes
```

```python
dados1.to_csv('pesquisa_tratada.csv', sep=',', encoding='UTF-8', index = False)
```
